```python
import math
a = float(input('Enter a: '))
b = float(input('Enter b: '))
c = float(input('Enter c: '))
d = (b**2) - (4*a*c)
sol1 = (-b-math.sqrt(d))/(2*a)
sol2 = (-b+math.sqrt(d))/(2*a)
sol1_is_zero = int(round(a*(sol1)**2 + b*(sol1) + c))
sol2_is_zero = int(round(a*(sol2)**2 + b*(sol2) + c))

if(sol1_is_zero == 0):
    print(sol1)
else:
    print(str(sol1)+' is not an answer')
if(sol2_is_zero == 0):
    print(sol2)
else:
    print(str(sol2) +' is not an answer')
```

```python
import math
from matplotlib import pyplot as plt
import numpy as np
N = 1000
def HC(T):
    K= 9*.001*6.022e28*1.380649e-23/(428**3)
    def f(x):
        if x == 0:
            return(0)
        else:
            return (K*T**3*(((x)**4)*(math.e**(x))/((math.e**(x))-1)**2))

    A = 0
    B = 428/T
    H = (B-A)/N
    S = .5*f(B)
    for n in range (1,N-1):
        S += f(n*H)
    return H*S*K
x=[]
y=[]
for T in range(5,501):
    x.append(T)
    y.append(HC(T))
print(x,y)

fig1 = plt.figure(figsize=(20,10))
fig1 = fig1.add_subplot(111)
fig1.scatter(x,y,s=20,c='g',alpha=0.9)
fig1.set_xlabel("$Temerature [K]$" , size=20)
fig1.set_ylabel("$Heat$" + " " + "$Capacity [J/K]$", size=20)
plt.title("Heat Capacity versus Temperature of Material", size = 30)
fig1.xaxis.set_tick_params(labelsize=20)
fig1.yaxis.set_tick_params(labelsize=20)

    
```
